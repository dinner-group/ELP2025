```python
### PLOTS EXPERIMENTAL DATA FOR FIG 1 IN MAIN TEXT ###
### 1 Aug 2025 ###

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
from pathlib import Path

# -----------------------------------------------------------------------------
# CONFIGURATION
# -----------------------------------------------------------------------------

# PATHS
BASE_DIR = Path('../data')
WAVENUMBERS_FILE = BASE_DIR / 'waxis.txt'
EXP_DIR = BASE_DIR / 'exp_data' / 'processed_FTIRs'

# DATA

# Isotopologues in the order they appear
ISOTOPES = ['WT', 'G3+', 'G5+', 'V1', 'G3', 'V4', 'G5+G3', 'V1V4', 'V1G3V4']

# Maps internal names to display names
ISOTOPE_LABELS = {
    'WT': 'UL', 'G3+': 'G1', 'G5+': 'G3', 'V1': 'V4', 'G3': 'G6', 
    'V4': 'V7', 'G5+G3': 'G3G6', 'V1V4': 'V4V7', 'V1G3V4': 'V4G6V7'
}

# Maps to exp. data files
ISOTOPE_FILE_MAP = {
    'WT': 'GVGn1_WT_50mgmL_phos_150mM_pH1_basecorr_processed.npy',
    'G3+': 'GVGn1_G3dag_50mgmL_phos_150mM_pH1_basecorr_processed.npy',
    'G5+': 'GVGn1_G5dag_50mgmL_phos_150mM_pH1_basecorr_processed.npy',
    'V1': 'GVGn1_V1_50mgmL_phos_150mM_pH1_basecorr_processed.npy',
    'G3': 'GVGn1_G3_50mgmL_phos_150mM_pH1_basecorr_processed.npy',
    'V4': 'GVGn1_V4_50mgmL_phos_150mM_pH1_basecorr_processed.npy',
    'G5+G3': 'GVGn1_G5dagG3_50mgmL_phos_150mM_pH1_basecorr_processed.npy',
    'V1V4': 'GVGn1_V1V4_50mgmL_phos_150mM_pH1_basecorr_processed.npy',
    'V1G3V4': 'GVGn1_V1G3V4_50mgmL_phos_150mM_pH1_basecorr_processed.npy'
}

# (label_text, x_position, y_offset) for annotating peaks
PEAK_LABELS = {
    'WT': [('UL', 1655, -0.5)],
    'G3+': [('G1', 1615, -0.25)],
    'G5+': [('G3', 1585, -0.17)],
    'V1': [('V4', 1555, -0.17)],
    'G3': [('G6', 1585, -0.15)],
    'V4': [('V7', 1585, -0.15)],
    'G5+G3': [('G3,G6', 1585, -0.22)],
    'V1V4': [('V4        V7', 1570, -0.18)], # space for visual clarity
    'V1G3V4': [('V4', 1555, -0.17), ('V7', 1585, -0.17), ('G6', 1610, -0.15)]
}

# Analysis Parameters
WAVENUMBER_MIN = 1525.0
WAVENUMBER_MAX = 1700.0
WAVENUMBER_FULL_MIN = 1525.0
WAVENUMBER_FULL_MAX = 1750.0
OFFSET_STEP = 0.45

# Plotting Specifications
FIGURE_SIZE = (3.25, 6)
DPI = 600

# -----------------------------------------------------------------------------
# DATA LOADING
# -----------------------------------------------------------------------------

# Filter to desired range
def load_wavenumbers():
    """Load and trim the wavenumber axis."""
    wavenumbers = np.loadtxt(WAVENUMBERS_FILE)
    wavenumbers_refined = wavenumbers[(wavenumbers >= WAVENUMBER_FULL_MIN) & 
                                     (wavenumbers <= WAVENUMBER_FULL_MAX)]
    mask = (wavenumbers_refined >= WAVENUMBER_MIN) & (wavenumbers_refined <= WAVENUMBER_MAX)
    return wavenumbers_refined[mask]

# Normalize by max wavenumber for visual purposes
def load_spectra_max_normalized(isotope_names):
    """Load spectra and normalize by max value."""
    spectra = []
    wavenumbers_full = np.loadtxt(WAVENUMBERS_FILE)
    wavenumbers_refined = wavenumbers_full[(wavenumbers_full >= WAVENUMBER_FULL_MIN) & 
                                          (wavenumbers_full <= WAVENUMBER_FULL_MAX)]
    mask = (wavenumbers_refined >= WAVENUMBER_MIN) & (wavenumbers_refined <= WAVENUMBER_MAX)

    for name in isotope_names:
        filepath = EXP_DIR / ISOTOPE_FILE_MAP[name]
        data = np.load(filepath)
        normalized = data / np.max(data)
        spectra.append(normalized[mask])
    return spectra

# -----------------------------------------------------------------------------
# ANALYSIS
# -----------------------------------------------------------------------------

# Used later to draw guide lines only in gaps between spectra
def find_segments(x, mask):
    """Find continuous segments where mask is True."""
    segments = []
    start = None
    for idx, val in enumerate(mask):
        if val and start is None:
            start = x[idx]
        elif not val and start is not None:
            segments.append((start, x[idx]))
            start = None
    if start is not None:
        segments.append((start, x[-1]))
    return segments

# -----------------------------------------------------------------------------
# PLOTTING
# -----------------------------------------------------------------------------

def set_plot_style():
    """Set consistent plot style for publication figures."""
    plt.rcParams.update({
        'font.family': 'Arial',
        'font.size': 10,  # Increase from 8
        'axes.linewidth': 1.0,
        'xtick.major.width': 1.0,
        'ytick.major.width': 1.0,
    })

def plot_peak_labels(ax, wavenumbers, spectrum, labels):
    """Plot peak annotation text."""
    for text, x, y_offset in labels:
        idx = np.argmin(np.abs(wavenumbers - x))
        y = spectrum[idx] + y_offset
        ax.text(x, y, text, fontsize=8, ha='center', va='bottom')

def add_guide_lines(ax, wavenumbers, spectra, idx):
    """Add horizontal guide lines between spectra."""
    y_offset = idx * OFFSET_STEP
    if idx == 0:
        ax.axhline(y=y_offset, color='lightgray', linewidth=0.5, alpha=0.7)
        return
    prev = spectra[idx - 1]
    prev_offset = (idx - 1) * OFFSET_STEP
    mask = prev + prev_offset < y_offset - 0.05
    for start, end in find_segments(wavenumbers, mask):
        ax.plot([start, end], [y_offset, y_offset], color='lightgray', linewidth=0.5, alpha=0.7)

def draw_label_box(ax, x, y, width, height, text):
    """Draw a single box for an isotope label."""
    rect = Rectangle((x - width/2, y - height/2), width, height,
                     linewidth=0.5, edgecolor='black', facecolor='white', alpha=0.9)
    ax.add_patch(rect)
    ax.text(x, y, text, fontsize=8, ha='center', va='center')

def add_sample_labels(ax, isotope_names, n_spectra):
    """Add all the isotope label boxes to the plot."""
    for idx, name in enumerate(isotope_names):
        label = ISOTOPE_LABELS.get(name, name)
        box_width = 40 if label == 'V4G6V7' else 20
        box_height = 0.15
        box_y = 8 * OFFSET_STEP + 1.25 - 0.1 if idx == n_spectra - 1 else (idx + 1) * OFFSET_STEP - 0.1
        box_x = 1543.0 if idx == n_spectra - 1 else 1535.0
        draw_label_box(ax, box_x, box_y, box_width, box_height, label)

def add_header_boxes(ax):
    """Add the gray header boxes at the top of the plot."""
    header_y = 8 * OFFSET_STEP + 1.25
    headers = [(1525, 85, 'labels'), (1610, 15, 'V4'), (1625, 75, 'main band')]
    for x, width, text in headers:
        rect = Rectangle((x, header_y), width, 0.4,
                         linewidth=1, edgecolor='black', facecolor='lightgray', alpha=0.8)
        ax.add_patch(rect)
        ax.text(x + width/2, header_y + 0.15, text, fontsize=8, ha='center', va='center')

def format_axes(ax, wavenumbers, spectra):
    """Set final axes properties."""
    ax.set_xlim([WAVENUMBER_MIN, WAVENUMBER_MAX])
    ax.set_xlabel(r'$\omega$ (cm$^{-1}$)', fontsize=8)
    ax.set_ylabel('Absorbance (arb. units)', fontsize=8)
    ax.set_yticks([])
    top_offset = (len(spectra) - 1) * OFFSET_STEP
    max_y = top_offset + max(spectra[-1]) + 0.5
    ax.set_ylim([-0.1, max_y])
    for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_linewidth(1)
        spine.set_color('black')

# -----------------------------------------------------------------------------
# MAIN
# -----------------------------------------------------------------------------

def main():
    """Main function to generate the plot."""
    set_plot_style()
    
    wavenumbers = load_wavenumbers()
    spectra = load_spectra_max_normalized(ISOTOPES)

    fig, ax = plt.subplots(figsize=FIGURE_SIZE, dpi=DPI)

    for idx, (spectrum, name) in enumerate(zip(spectra, ISOTOPES)):
        y_offset = idx * OFFSET_STEP
        ax.plot(wavenumbers, spectrum + y_offset, linewidth=1, label=name)
        if name in PEAK_LABELS:
            plot_peak_labels(ax, wavenumbers, spectrum + y_offset, PEAK_LABELS[name])
        add_guide_lines(ax, wavenumbers, spectra, idx)

    add_header_boxes(ax)
    add_sample_labels(ax, ISOTOPES, len(spectra))
    format_axes(ax, wavenumbers, spectra)

    plt.subplots_adjust(left=0.15, top=0.95, right=0.98, bottom=0.15)
    plt.show()

if __name__ == "__main__":
    main()
```
