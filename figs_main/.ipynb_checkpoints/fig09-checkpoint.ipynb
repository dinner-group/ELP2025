```python
### PLOTS FIG 9 IN MAIN TEXT ###
### 2 Aug 2025 ###

import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# -----------------------------------------------------------------------------
# CONFIGURATION
# -----------------------------------------------------------------------------

# Paths
BASE_DIR = Path('../data')
ERROR_DIR = BASE_DIR / 'error_data' / 'overlap_informative'
PROP_DIR = BASE_DIR / 'prop_array'

# Data Specifications
FORCE_FIELDS = ['CHARMM27_TIP3P', 'CHARMM36m_TIP3P']
FORCE_FIELD_LABELS = {'CHARMM27_TIP3P': 'C27', 'CHARMM36m_TIP3P': 'C36m'}
ISOTOPES = ['G5+', 'V1', 'V4', 'G5+G3', 'V1V4', 'V1G3V4']
ISOTOPE_LABELS = {
    'G5+': 'G3', 'V1': 'V4', 'V4': 'V7',
    'G5+G3': 'G3G6', 'V1V4': 'V4V7', 'V1G3V4': 'V4G6V7'
}

# State definitions and analysis parameters
STRUCTURE_TYPES = {
    'G5+': 'N', 'V1': 'C', 'V4': 'C',
    'G5+G3': 'N', 'V1V4': 'C', 'V1G3V4': 'C'
}
STATE_COLORS = {
    'C': ['#8c564b', '#17becf'],
    'N': ['#9467bd', '#e377c2']
}
STATE_KEYS = ['C', 'E', 'NC', 'N']

# Analysis Parameters
FORCE_FIELD_HATCHES = {'CHARMM27_TIP3P': '', 'CHARMM36m_TIP3P': '///'}
FORCE_FIELD_STYLES = {'CHARMM27_TIP3P': '--', 'CHARMM36m_TIP3P': ':'}
ESS_TARGETS = {'CHARMM27_TIP3P': 100, 'CHARMM36m_TIP3P': 1000}

# Plotting Specifications
FIGURE_SIZE = (3.25, 3.8)
DPI = 600

# -----------------------------------------------------------------------------
# DATA LOADING
# -----------------------------------------------------------------------------

def load_error_data(isotope, force_field, suffix):
    """Load error data for given isotope and force field."""
    filename = f'error_overlap_{isotope}_{force_field}_ch27_0_{suffix}.npy'
    filepath = ERROR_DIR / filename
    # SHAPE: (5000, 1000_values) - typically something like (1000, 100)
    return np.load(filepath) if filepath.exists() else None

def load_proportions(force_field):
    """Load proportions for given force field."""
    filename = f'proportions_array_0100ps_{force_field}_PAPER.npy'
    filepath = PROP_DIR / force_field / filename
    # SHAPE: (5000, 4) - 4 states: ['C', 'E', 'NC', 'N']
    return np.load(filepath) if filepath.exists() else None

# -----------------------------------------------------------------------------
# ANALYSIS
# -----------------------------------------------------------------------------

def compute_weights(error):
    """Compute normalized weights from error."""
    # INPUT error shape: (5000, 1000)
    weights = np.exp(-error)  # Element-wise exponential
    # Normalize each column (each theta value) to sum to 1
    weights /= np.sum(weights, axis=0, keepdims=True)
    # OUTPUT weights shape: (5000, 1000) - each column sums to 1
    return weights

def compute_ess(weights):
    """Compute effective sample size from weights."""
    # INPUT weights shape: (5000, 1000)
    weight_sum = np.sum(weights, axis=0)  # Shape: (1000,)
    weight_sum_squared = np.sum(weights**2, axis=0)  # Shape: (1000,)
    # ESS = (sum of weights)^2 / (sum of squared weights)
    # OUTPUT shape: (1000,) - one ESS value per theta
    return np.divide(weight_sum**2, weight_sum_squared, 
                    out=np.zeros_like(weight_sum), 
                    where=weight_sum_squared != 0)

def find_best_theta_index(ess, target):
    """Find index corresponding to target ESS."""
    # INPUT ess shape: (1000,)
    if np.nanmax(ess) < target:
        return None
    valid = ~np.isnan(ess)
    # Find theta index where ESS is closest to target
    return np.argmin(np.abs(ess - target) + (~valid * 1e6))

def combine_population(populations, structure_type):
    """Combine populations based on structure type."""
    # INPUT populations shape: (4,) - [C, E, NC, N]
    if structure_type == 'C':
        # For C-turn analysis:
        has_feature = populations[0] + populations[2]  # C + NC (has C-turn)
        no_feature = populations[1] + populations[3]   # E + N (no C-turn)
    else:  # 'N'
        # For N-bend analysis:
        has_feature = populations[2] + populations[3]  # NC + N (has N-bend)
        no_feature = populations[0] + populations[1]   # C + E (no N-bend)
    # OUTPUT shape: (2,) - [has_feature, no_feature]
    return np.array([has_feature, no_feature]) # should sum to 1

def compute_weighted_population(error, proportions, target_ess, structure_type):
    """Compute weighted population at target ESS."""
    # INPUT error shape: (5000, 1000)
    # INPUT proportions shape: (5000, 4)
    
    weights = compute_weights(error)  # Shape: (5000, 1000)
    ess = compute_ess(weights)  # Shape: (1000,)
    idx = find_best_theta_index(ess, target_ess)  # Single index
    
    if idx is None:
        return None, None
    
    # Extract weights for the best theta
    weight_vector = weights[:, idx]  # Shape: (5000,)
    
    # Weighted average of proportions
    # proportions: (5000, 4), weight_vector: (5000,)
    # weight_vector[:, None] makes it (5000, 1) for broadcasting for the weighted sum
    mean_population = np.sum(proportions * weight_vector[:, None], axis=0)  # Shape: (4,)
    
    # Combine into 2 categories
    combined = combine_population(mean_population, structure_type)  # Shape: (2,)
    
    return combined, ess[idx]

def compute_unweighted_population(proportions, structure_type):
    """Compute unweighted average population."""
    # INPUT proportions shape: (5000, 4)
    mean_population = np.mean(proportions, axis=0)  # Shape: (4,)
    # OUTPUT shape: (2,)
    return combine_population(mean_population, structure_type)

# -----------------------------------------------------------------------------
# PLOTTING
# -----------------------------------------------------------------------------

def set_plot_style():
    """Set consistent plot style for publication figures."""
    plt.rcParams.update({
        'font.family': 'Arial',
        'font.size': 10,
        'axes.linewidth': 1.0,
        'xtick.major.width': 1.0,
        'ytick.major.width': 1.0,
        'lines.linewidth': 1.0,
        'figure.dpi': DPI
    })

def plot_bar(ax, x, population, color_yes, color_no, hatch, width=0.35):
    """Plot stacked bar for population."""
    # INPUT population shape: (2,) - [has_feature, no_feature]
    # x: horizontal position (0 for first force field, 0.4 for second)
    
    # Bottom bar: no_feature population
    ax.bar(x, population[1], width, color=color_no, hatch=hatch, alpha=0.8)
    
    # Top bar: has_feature population (stacked on top)
    ax.bar(x, population[0], width, bottom=population[1], color=color_yes, hatch=hatch, alpha=0.8)
    
    # Total height = population[0] + population[1] = 1.0

def add_reference_lines(ax, population, color_yes, color_no, style):
    """Add reference lines for original populations."""
    # Horizontal line at height of no_feature population
    ax.axhline(y=population[1], color=color_no, linestyle=style, linewidth=1.5, alpha=0.7)

def format_axes(ax, title, show_ylabel):
    """Format subplot axes."""
    ax.set_xlim(-0.1, 0.5)  # Two bars at x=0 and x=0.4
    ax.set_ylim(0, 1)       # Population fraction from 0 to 1
    ax.set_xticks([0, 0.4]) # Positions of the two bars
    ax.set_xticklabels([FORCE_FIELD_LABELS[force_field] for force_field in FORCE_FIELDS], 
                      fontsize=8)
    ax.set_title(title, fontsize=12)
    if show_ylabel:
        ax.set_ylabel('Population', fontsize=10)

def create_legend():
    """Create legend elements."""
    from matplotlib.patches import Rectangle
    from matplotlib.lines import Line2D
    
    elements = [
        Rectangle((0,0),1,1,fc=STATE_COLORS['N'][0], alpha=0.8, label='N-bend'),
        Rectangle((0,0),1,1,fc=STATE_COLORS['N'][1], alpha=0.8, label='no N-bend'),
        Rectangle((0,0),1,1,fc=STATE_COLORS['C'][0], alpha=0.8, label='C-turn'),
        Rectangle((0,0),1,1,fc=STATE_COLORS['C'][1], alpha=0.8, label='no C-turn'),
        Rectangle((0,0),1,1,fc='gray', hatch=FORCE_FIELD_HATCHES['CHARMM27_TIP3P'], 
                 alpha=0.8, label='C27'),
        Rectangle((0,0),1,1,fc='gray', hatch=FORCE_FIELD_HATCHES['CHARMM36m_TIP3P'], 
                 alpha=0.8, label='C36m'),
        Line2D([0], [0], color='black', linestyle=FORCE_FIELD_STYLES['CHARMM27_TIP3P'], 
               lw=2, label='C27 original'),
        Line2D([0], [0], color='black', linestyle=FORCE_FIELD_STYLES['CHARMM36m_TIP3P'], 
               lw=2, label='C36m original'),
    ]
    return elements

# -----------------------------------------------------------------------------
# MAIN
# -----------------------------------------------------------------------------

def main():
    """Main function to generate the plot."""
    set_plot_style()
    
    # Create figure with 2x3 subplots
    # LAYOUT: 
    # [0,0]: G3 (N-bend)    [0,1]: V4 (C-turn)    [0,2]: V7 (C-turn)
    # [1,0]: G3G6 (N-bend)  [1,1]: V4V7 (C-turn)  [1,2]: V4G6V7 (C-turn)
    
    fig, axes = plt.subplots(2, 3, figsize=FIGURE_SIZE, sharey=True, dpi=DPI)
    fig.subplots_adjust(top=0.95, bottom=0.20, hspace=0.35, wspace=0.25)

    # Process each isotope (6 total)
    for idx, isotope in enumerate(ISOTOPES):
        structure_type = STRUCTURE_TYPES[isotope]  # 'C' or 'N'
        colors = STATE_COLORS[structure_type]      # [color_yes, color_no]
        
        # Calculate subplot position
        row, col = idx // 3, idx % 3  # Examples - idx=0: (0,0), idx=1: (0,1), idx=4: (1,1)
        ax = axes[row, col]
        
        print(f"\n{'='*60}")
        print(f"Isotope: {isotope} ({ISOTOPE_LABELS[isotope]}) - Structure type: {structure_type}")
        print(f"Subplot position: row={row}, col={col}")
        
        # Process each force field (2 total per subplot)
        for force_field_idx, force_field in enumerate(FORCE_FIELDS):
            # Load data
            proportions = load_proportions(force_field)
            error = load_error_data(isotope, force_field, structure_type)
            
            if proportions is None or error is None:
                print(f"ERROR:  Missing data for {force_field}")
                continue
            
            print(f"\n  Force field: {force_field}")
            print(f"  Error shape: {error.shape}")
            print(f"  Proportions shape: {proportions.shape}")
            
            # Compute populations
            target = ESS_TARGETS[force_field]
            weighted_population, actual_ess = compute_weighted_population(
                error, proportions, target, structure_type
            )
            
            if weighted_population is None:
                print(f"  Could not achieve target ESS of {target}")
                continue
            
            original_population = compute_unweighted_population(proportions, structure_type) # FF prior
            
            # Print population values for Figure 10
            feature_name = "N-bend" if structure_type == 'N' else "C-turn"
            print(f"  Weighted population (ESS={actual_ess:.1f}):")
            print(f"    Has {feature_name}: {weighted_population[0]:.3f}")
            print(f"    No {feature_name}: {weighted_population[1]:.3f}")
            print(f"  Original (unweighted) population:")
            print(f"    Has {feature_name}: {original_population[0]:.3f}")
            print(f"    No {feature_name}: {original_population[1]:.3f}")
            
            # Plot at x=0 for CHARMM27, x=0.4 for CHARMM36m
            x_position = force_field_idx * 0.4
            
            # Plot stacked bar
            plot_bar(ax, x_position, weighted_population, 
                    colors[0], colors[1], FORCE_FIELD_HATCHES[force_field])
            
            # Add reference lines showing original populations
            add_reference_lines(ax, original_population, colors[0], colors[1], 
                              FORCE_FIELD_STYLES[force_field])
        
        # Format this subplot
        format_axes(ax, ISOTOPE_LABELS[isotope], show_ylabel=(idx % 3 == 0))

    # Add legend below all subplots
    legend_elements = create_legend()
    fig.legend(handles=legend_elements, loc='upper center', ncol=4, fontsize=8,
               bbox_to_anchor=(0.5, 0.05), columnspacing=0.5, handlelength=1.2,
               handletextpad=0.4, frameon=False)
    
    plt.tight_layout()
    plt.subplots_adjust(bottom=0.15)  # Make room for legend
    plt.show()

if __name__ == "__main__":
    main()
```
