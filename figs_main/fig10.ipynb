```python
### PLOTS SOLUTION TO SYSTEM OF EQUATIONS (FIG 10) IN MAIN TEXT ###
### 2 Aug 2025 ###

import numpy as np
import matplotlib.pyplot as plt

# -----------------------------------------------------------------------------
# CONFIGURATION
# -----------------------------------------------------------------------------

# Data Specifications
FORCE_FIELD_LABELS = ['CHARMM27', 'CHARMM36m']
P_E_PLUS_N_VALUES = [0.726, 0.940]  # p(E) + p(N) for ['CHARMM27', 'CHARMM36m']
P_E_PLUS_C_VALUES = [0.717, 0.976]  # p(E) + p(C) for ['CHARMM27', 'CHARMM36m']

# State definitions
STATE_KEYS = ['E', 'N', 'C', 'NC']
STATE_COLORS = {
    'E': 'orange',
    'N': 'green',
    'C': 'blue',
    'NC': 'red'
}
STATE_LINESTYLES = {
    'E': '-',
    'N': '--',
    'C': '-.',
    'NC': ':'
}

# Analysis Parameters
NC_POPULATION_RANGE = np.linspace(0, 0.03, 1000)

# Plotting Specifications
FIGURE_SIZE = (3.25, 2.5)
DPI = 600

# -----------------------------------------------------------------------------
# ANALYSIS
# -----------------------------------------------------------------------------

def compute_populations(p_nc, p_e_plus_n, p_e_plus_c):
    """Calculate populations of E, N, and C based on constraints.
    Solution using p_NC as free parameter
    
    Args:
        p_nc: Population of NC state
        p_e_plus_n: Constraint value for p(E) + p(N)
        p_e_plus_c: Constraint value for p(E) + p(C)
        
    Returns:
        Dictionary with populations for each state
    """
    p_e = p_nc + p_e_plus_n + p_e_plus_c - 1
    p_n = 1 - p_e_plus_c - p_nc
    p_c = 1 - p_e_plus_n - p_nc
    return {'E': p_e, 'N': p_n, 'C': p_c, 'NC': p_nc}



# -----------------------------------------------------------------------------
# MAIN
# -----------------------------------------------------------------------------

def main():
    """Main function to generate the plot."""
    print("="*60)
    print("POPULATION CONSTRAINTS ANALYSIS")
    print("="*60)
    
    # Set consistent plot style for publication figures
    plt.rcParams.update({
        'font.family': 'Arial',
        'font.size': 10,
        'axes.linewidth': 1.0,
        'xtick.major.width': 1.0,
        'ytick.major.width': 1.0,
        'lines.linewidth': 1.0,
        'figure.dpi': DPI
    })
    
    # Create the main figure and arrange subplots
    fig, axes = plt.subplots(1, 2, figsize=FIGURE_SIZE, sharey=True)
    
    # Plot each force field
    for idx, (label, p_e_plus_n, p_e_plus_c) in enumerate(zip(
            FORCE_FIELD_LABELS, P_E_PLUS_N_VALUES, P_E_PLUS_C_VALUES)):
        
        ax = axes[idx]
        
        # Compute populations across NC range
        populations = compute_populations(NC_POPULATION_RANGE, p_e_plus_n, p_e_plus_c)
        
        # Print population shapes
        print(f"\n{label} population shapes:")
        for state in STATE_KEYS:
            print(f"  {state}: {populations[state].shape}")
        
        # Plot each state (three lines)
        for state in STATE_KEYS:
            ax.plot(NC_POPULATION_RANGE, populations[state], 
                    color=STATE_COLORS[state], 
                    linestyle=STATE_LINESTYLES[state], 
                    label=state)
        
        # Format axes for single subplot
        ax.set_xlim(0, 0.03)
        ax.set_ylim(0, 1)
        ax.set_xlabel('$p_{NC}$')
        ax.set_title(f'{label}\n$p_{{E}} + p_{{N}}$ = {p_e_plus_n:.3f}\n$p_{{E}} + p_{{C}}$ = {p_e_plus_c:.3f}', 
                     fontsize=9)
        ax.tick_params(axis='both', which='major', labelsize=8)
        if idx == 0:  # show_ylabel for first subplot
            ax.set_ylabel('Population')
    
    # Adjust layout and add legend
    plt.tight_layout()
    plt.subplots_adjust(wspace=0.3, bottom=0.25)
    
    fig.legend(STATE_KEYS, loc='lower center', bbox_to_anchor=(0.5, -0.05), 
               ncol=4, fontsize=8, frameon=False)
    
    plt.show()
    
    print("\n" + "="*60)
    print("POPULATION CONSTRAINTS ANALYSIS COMPLETE!")
    print("="*60)

if __name__ == "__main__":
    main()
```
