```python
### PLOTS SPECTRAL ENSEMBLES FOR FIG 2 IN MAIN TEXT ###
### 1 Aug 2025 ###

import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import warnings

# -----------------------------------------------------------------------------
# CONFIGURATION
# -----------------------------------------------------------------------------

# PATHS
BASE_DIR = Path('../data')
EXP_DIR = BASE_DIR / 'exp_data' / 'processed_FTIRs'
SIM_DIR = BASE_DIR / 'sim_data' / 'sim_spec'
WAVENUMBERS_FILE = BASE_DIR / 'waxis.txt'

# DATA
ISOTOPES = ['WT', 'G5+', 'V1G3V4']
FORCE_FIELDS = ['CHARMM27_TIP3P', 'CHARMM36m_TIP3P']
MAP_TYPE = 'DC1Fnew'

ISOTOPE_LABELS = {'WT': 'UL', 'G5+': 'G3', 'V1G3V4': 'V4G6V7'}
FORCE_FIELD_LABELS = {
    'CHARMM27_TIP3P': 'CHARMM27',
    'CHARMM36m_TIP3P': 'CHARMM36m'
}
ISOTOPE_FILE_MAP = {
    'WT': 'GVGn1_WT_50mgmL_phos_150mM_pH1_basecorr_processed.npy',
    'G5+': 'GVGn1_G5dag_50mgmL_phos_150mM_pH1_basecorr_processed.npy',
    'V1G3V4': 'GVGn1_V1G3V4_50mgmL_phos_150mM_pH1_basecorr_processed.npy'
}

# Analysis Parameters
WAVENUMBER_MIN_ANALYSIS = 1525
WAVENUMBER_MAX_ANALYSIS = 1750
WAVENUMBER_MIN_PLOT = 1525
WAVENUMBER_MAX_PLOT = 1710

# Plotting Specifications
FIGURE_SIZE = (3.25, 4.5)
DPI = 600

# -----------------------------------------------------------------------------
# DATA LOADING
# -----------------------------------------------------------------------------

def load_wavenumbers(wavenumber_min, wavenumber_max):
    """Load wavenumbers within specified range."""
    wavenumbers = np.loadtxt(WAVENUMBERS_FILE)
    return wavenumbers[(wavenumbers >= wavenumber_min) & (wavenumbers <= wavenumber_max)]

def load_experimental_spectrum(isotope, wavenumbers_full):
    """Load and normalize experimental spectrum."""
    filepath = EXP_DIR / ISOTOPE_FILE_MAP[isotope]
    intensities = np.load(filepath)
    normalized_intensities = normalize_by_area(intensities, wavenumbers_full)
    return trim_spectrum(wavenumbers_full, normalized_intensities, 
                        WAVENUMBER_MIN_PLOT, WAVENUMBER_MAX_PLOT)

def load_simulation_spectra(isotope, force_field, wavenumbers_full):
    """Load and normalize all simulation spectra for given isotope and force field."""
    filename = f"{MAP_TYPE}_{force_field}_{isotope}_ftir.npy"
    filepath = SIM_DIR / force_field / filename
    
    if not filepath.exists():
        warnings.warn(f"Missing simulation file: {filepath}")
        return []
    
    sim_spectra = []
    raw_data = np.load(filepath)
    
    for row in raw_data:
        corrected = row - np.min(row)
        normalized = normalize_by_area(corrected, wavenumbers_full)
        _, trimmed_intensities = trim_spectrum(wavenumbers_full, normalized, 
                                             WAVENUMBER_MIN_PLOT, WAVENUMBER_MAX_PLOT)
        sim_spectra.append(trimmed_intensities)
    
    return sim_spectra

# -----------------------------------------------------------------------------
# ANALYSIS
# -----------------------------------------------------------------------------

def normalize_by_area(intensities, wavenumbers):
    """Normalize spectrum by integrated area."""
    area = np.trapz(intensities, x=wavenumbers)
    return intensities / area if area > 0 else intensities

def trim_spectrum(wavenumbers, intensities, wavenumber_min, wavenumber_max):
    """Trim spectrum to specified wavenumber range."""
    mask = (wavenumbers >= wavenumber_min) & (wavenumbers <= wavenumber_max)
    return wavenumbers[mask], intensities[mask]

# -----------------------------------------------------------------------------
# PLOTTING
# -----------------------------------------------------------------------------

def set_plot_style():
    """Set consistent plot style for ACS publication figures."""
    plt.rcParams.update({
        # Font settings - Arial (PC) or Helvetica (Mac) as per ACS
        'font.family': 'sans-serif',
        'font.sans-serif': ['Arial', 'Helvetica'],
        'font.size': 10,  # Will be readable at 5pt when reduced to column width
        
        # Line widths - minimum 1 pt as per ACS
        'lines.linewidth': 1.0,
        'axes.linewidth': 1.0,
        'xtick.major.width': 1.0,
        'ytick.major.width': 1.0,
        'xtick.minor.width': 1.0,
        'ytick.minor.width': 1.0,
        'patch.linewidth': 1.0,
        'grid.linewidth': 1.0,
        
        # Font sizes - accounting for reduction
        'axes.labelsize': 11,
        'axes.titlesize': 12,
        'xtick.labelsize': 10,
        'ytick.labelsize': 10,
        'legend.fontsize': 10,
        
        # DPI settings
        'figure.dpi': 600
    })


def plot_spectra(ax, wavenumbers, sim_spectra, exp_spectrum, color):
    """Plot simulation and experimental spectra."""
    # Plot individual simulations
    if sim_spectra:
        for sim_trace in sim_spectra:
            ax.plot(wavenumbers, sim_trace, color=color, linewidth=0.225, alpha=0.7)
        
        # Plot average simulation
        avg_sim = np.mean(sim_spectra, axis=0)
        ax.plot(wavenumbers, avg_sim, color='black', linewidth=1.5, linestyle=':')
    
    # Plot experimental
    ax.plot(wavenumbers, exp_spectrum, color='black', linewidth=1.5)

def format_subplot(ax, isotope, force_field, row_idx, col_idx):
    """Format individual subplot."""
    ax.set_xlim(WAVENUMBER_MIN_PLOT, WAVENUMBER_MAX_PLOT)
    ax.set_yticks([])
    
    if row_idx == 0:
        ax.set_title(FORCE_FIELD_LABELS[force_field], fontsize=10, pad=8)
    
    if col_idx == 0:
        ax.text(0.05, 0.95, ISOTOPE_LABELS[isotope], transform=ax.transAxes, 
                fontsize=10, va='top', ha='left',
                bbox=dict(boxstyle='round,pad=0.3', fc='white', alpha=0.8, ec='none'))
    
    ax.tick_params(axis='x', labelsize=8)

# -----------------------------------------------------------------------------
# MAIN
# -----------------------------------------------------------------------------

def main():
    """Main function to generate the plot."""
    set_plot_style()
    
    # Create figure
    fig, axes = plt.subplots(
        len(ISOTOPES), len(FORCE_FIELDS), 
        figsize=FIGURE_SIZE, sharex=True, sharey=True
    )
    plt.subplots_adjust(left=0.15, bottom=0.1, right=0.95, top=0.9, wspace=0, hspace=0)

    # Load data and create plots
    wavenumbers_full = load_wavenumbers(WAVENUMBER_MIN_ANALYSIS, WAVENUMBER_MAX_ANALYSIS)
    wavenumbers_plot, _ = trim_spectrum(wavenumbers_full, np.zeros_like(wavenumbers_full), 
                                       WAVENUMBER_MIN_PLOT, WAVENUMBER_MAX_PLOT)
    colors = plt.cm.rainbow(np.linspace(1, 0, len(ISOTOPES)))

    for row_idx, isotope in enumerate(ISOTOPES):
        for col_idx, force_field in enumerate(FORCE_FIELDS):
            ax = axes[row_idx, col_idx]
            
            # Load data
            sim_spectra = load_simulation_spectra(isotope, force_field, wavenumbers_full)
            _, exp_spectrum = load_experimental_spectrum(isotope, wavenumbers_full)
            
            # Plot spectra
            plot_spectra(ax, wavenumbers_plot, sim_spectra, exp_spectrum, colors[row_idx])
            
            # Format subplot
            format_subplot(ax, isotope, force_field, row_idx, col_idx)

    # Add shared labels
    fig.text(0.5, 0.03, r'$\omega$ (cm$^{-1}$)', ha='center', fontsize=10)
    fig.text(0.04, 0.5, r'Simulated Absorbance, $I(\omega) \propto \mathrm{Im}[\mathrm{FT}\{R(t)\}]$', 
         va='center', rotation='vertical', fontsize=10)
    
    plt.show()

if __name__ == "__main__":
    main()
```
