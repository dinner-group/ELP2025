```python
### PLOTS CORRESPONDENCE BETWEEN 4-STATE MODEL AND 16-STATE MODEL (FIG S07)
### 2 Aug 2025 ###

import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# -----------------------------------------------------------------------------
# CONFIGURATION
# -----------------------------------------------------------------------------

# Paths
BASE_DIR = Path('../data')
PROP_DIR = BASE_DIR / 'prop_array'

# Data Specifications
FORCE_FIELDS = ['CHARMM27_TIP3P', 'CHARMM36m_TIP3P']
FORCE_FIELD_LABELS = {
    'CHARMM27_TIP3P': 'CHARMM27',
    'CHARMM36m_TIP3P': 'CHARMM36m'
}

# State definitions
BINARY_LABELS = [f'{i:04b}' for i in range(16)]
STATE_4_TO_LABEL = {2: 'E', 4: 'N', 1: 'C', 3: 'NC'}
STATE_4_PLOT_ORDER = [2, 4, 1, 3]  # Corresponds to E, N, C, NC

# Plotting Specifications
FIGURE_SIZE = (7, 9)
DPI = 600
BAR_ALPHA = 0.8
MIN_PERCENTAGE_LABEL = 0.5

# -----------------------------------------------------------------------------
# DATA LOADING
# -----------------------------------------------------------------------------

def load_state_arrays(force_field):
    """Load 4-state and 16-state arrays for given force field."""
    data_path = PROP_DIR / force_field
    
    states_4_file = data_path / f'04_states_array_{force_field}.npy'
    states_16_file = data_path / f'16_states_array_{force_field}.npy'
    
    if not states_4_file.exists():
        raise FileNotFoundError(f'Missing 4-state file: {states_4_file}')
    if not states_16_file.exists():
        raise FileNotFoundError(f'Missing 16-state file: {states_16_file}')
    
    states_4 = np.load(states_4_file)
    states_16 = np.load(states_16_file)
    
    print(f'Loaded {force_field} data:')
    print(f'  4-state array: {states_4.shape}')
    print(f'  16-state array: {states_16.shape}')
    
    return states_4, states_16

# -----------------------------------------------------------------------------
# ANALYSIS
# -----------------------------------------------------------------------------

def calculate_state_fractions(state_data):
    """Calculate fractional population of each 16-state."""
    state_counts = np.zeros(16)
    unique_states, unique_counts = np.unique(state_data, return_counts=True)
    state_counts[unique_states] = unique_counts
    fractions = state_counts / len(state_data)
    
    return fractions

def print_state_statistics(force_field_label, subset_label, data, fractions):
    """Print comprehensive statistics for state distribution."""
    print(f'\n{force_field_label} - {subset_label}:')
    print(f'  Total frames: {len(data):,}')
    
    non_zero_states = np.sum(fractions > 0)
    print(f'  Populated states: {non_zero_states}/16')
    
    # Print details for populated states
    for state_index, fraction in enumerate(fractions):
        if fraction > 0:
            percentage = fraction * 100
            binary_label = BINARY_LABELS[state_index]
            print(f'    State {state_index:2d} ({binary_label}): {percentage:6.2f}%')
    
    print('-' * 60)

def extract_subset_data(states_4, states_16, target_4_state):
    """Extract 16-state data for frames in specific 4-state."""
    indices = np.where(states_4 == target_4_state)[0]
    subset_16_states = states_16[indices]
    
    return subset_16_states, indices

# -----------------------------------------------------------------------------
# PLOTTING
# -----------------------------------------------------------------------------

def set_plot_style():
    """Set consistent plot style for ACS publication figures."""
    plt.rcParams.update({
        # Font settings - Arial (PC) or Helvetica (Mac) as per ACS
        'font.family': 'sans-serif',
        'font.sans-serif': ['Arial', 'Helvetica'],
        'font.size': 10,  # Will be readable at 5pt when reduced to column width
        
        # Line widths - minimum 1 pt as per ACS
        'lines.linewidth': 1.0,
        'axes.linewidth': 1.0,
        'xtick.major.width': 1.0,
        'ytick.major.width': 1.0,
        'xtick.minor.width': 1.0,
        'ytick.minor.width': 1.0,
        'patch.linewidth': 1.0,
        'grid.linewidth': 1.0,
        
        # Font sizes - accounting for reduction
        'axes.labelsize': 11,
        'axes.titlesize': 12,
        'xtick.labelsize': 10,
        'ytick.labelsize': 10,
        'legend.fontsize': 10,
        
        # DPI settings
        'figure.dpi': 600
    })

def create_state_colormap():
    """Create color mapping for 16 states using rainbow colormap."""
    colormap = plt.cm.rainbow(np.linspace(0, 1, 16))
    return [colormap[i] for i in range(16)]

def add_percentage_labels_to_bars(ax, bars, fractions, column_index):
    """Add percentage labels above bars exceeding minimum threshold."""
    for state_index, (bar, fraction) in enumerate(zip(bars, fractions)):
        percentage = fraction * 100
        
        if percentage >= MIN_PERCENTAGE_LABEL:
            # Adjust x-position for overlapping labels
            x_offset = 0.0
            if column_index == 0 and state_index in [7, 15]:
                x_offset = bar.get_width() * 0.3
            
            label_x = bar.get_x() + bar.get_width() / 2 + x_offset
            label_y = bar.get_height() + 0.01
            
            ax.text(label_x, label_y, f'{percentage:.1f}%',
                   ha='center', va='bottom', fontsize=8)

def plot_state_distribution(ax, fractions, colors, column_index, row_index, 
                          force_field_label, y_label=None):
    """Plot state distribution for single panel."""
    # Create bar plot
    bars = ax.bar(range(16), fractions, color=colors, alpha=BAR_ALPHA)
    
    # Set axis properties
    ax.set_ylim(0, 1)
    ax.set_xticks(range(16))
    ax.set_xticklabels(BINARY_LABELS, rotation=45, fontsize=8, ha='right')
    ax.set_yticklabels([])
    ax.tick_params(axis='both', labelsize=8, width=1.0)
    
    # Add title for top row
    if row_index == 0:
        ax.set_title(force_field_label, fontsize=12, pad=15)
    
    # Add x-label for bottom row
    if row_index == 4:
        ax.set_xlabel('State (Binary)', fontsize=11)
    
    # Add y-label for left column
    if column_index == 0:
        if row_index == 0:
            ax.set_ylabel('All', fontsize=11)
        elif y_label:
            ax.set_ylabel(y_label, fontsize=11)
    
    # Add percentage labels
    add_percentage_labels_to_bars(ax, bars, fractions, column_index)

def create_state_analysis_figure(all_data):
    """Create complete 5x2 figure with state distributions."""
    fig, axes = plt.subplots(5, 2, figsize=FIGURE_SIZE, dpi=DPI)
    
    # Create colors for 16 states
    state_colors = create_state_colormap()
    
    for column_index, force_field in enumerate(FORCE_FIELDS):
        states_4, states_16 = all_data[force_field]
        force_field_label = FORCE_FIELD_LABELS[force_field]
        
        # Plot overall distribution (row 0)
        fractions_all = calculate_state_fractions(states_16)
        plot_state_distribution(
            axes[0, column_index], fractions_all, state_colors,
            column_index, 0, force_field_label
        )
        
        # Print statistics for overall distribution
        print_state_statistics(force_field_label, 'All States', 
                             states_16, fractions_all)
        
        # Plot distributions for each 4-state subset (rows 1-4)
        for subset_index, state_4_value in enumerate(STATE_4_PLOT_ORDER): # STATE_4_PLOT_ORDER = [2, 4, 1, 3]  corresponding to E, N, C, NC
            row_index = subset_index + 1
            
            # Extract subset data
            subset_16_states, subset_indices = extract_subset_data(
                states_4, states_16, state_4_value
            )
            
            # Calculate fractions for subset (each 4-state)
            fractions_subset = calculate_state_fractions(subset_16_states)
            state_label = STATE_4_TO_LABEL[state_4_value]
            
            # Plot subset distribution within each state
            plot_state_distribution(
                axes[row_index, column_index], fractions_subset, state_colors,
                column_index, row_index, force_field_label, state_label
            )
            
            # Print statistics for subset
            subset_description = f'State "{state_label}" (4-state = {state_4_value})'
            print_state_statistics(force_field_label, subset_description,
                                 subset_16_states, fractions_subset)
    
    return fig

# -----------------------------------------------------------------------------
# MAIN
# -----------------------------------------------------------------------------

def main():
    """Main function to run state distribution analysis."""
    set_plot_style()
    
    print('State Distribution Analysis (4-state vs 16-state)')
    print('=' * 60)
    
    try:
        # Load all data
        all_data = {}
        for force_field in FORCE_FIELDS:
            states_4, states_16 = load_state_arrays(force_field)
            all_data[force_field] = (states_4, states_16)
        
        # Create analysis figure
        print('\nAnalyzing state distributions...')
        fig = create_state_analysis_figure(all_data)
        
        # Adjust layout with increased row spacing
        plt.tight_layout()
        plt.subplots_adjust(hspace=0.35, wspace=0.25, top=0.95, 
                           bottom=0.06, left=0.08, right=0.95)
        
        # Display figure
        plt.show()
        
        print('\nState distribution analysis completed successfully!')
        print('=' * 60)
        
    except Exception as e:
        print(f'Error during analysis: {e}')
        raise

if __name__ == '__main__':
    main()
```
