```python
### KERNEL DENSITY ESTIMATIONS (KDEs) FOR L-VALUES (FIG S11) ###
### 2 Aug 2025 ###

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
from pathlib import Path

# -----------------------------------------------------------------------------
# CONFIGURATION
# -----------------------------------------------------------------------------

# Paths
BASE_DIR = Path('../data')
ERROR_DIR = BASE_DIR / 'error_data' / 'overlap_full'

# Data Specifications
FORCE_FIELDS = ['CHARMM27_TIP3P', 'CHARMM36m_TIP3P']
FORCE_FIELD_LABELS = {
    'CHARMM27_TIP3P': 'CHARMM27/TIP3P',
    'CHARMM36m_TIP3P': 'CHARMM36m/TIP3P'
}

ISOTOPES = ['WT', 'G3+', 'G5+', 'V1', 'G3', 'V4', 'G5+G3', 'V1V4', 'V1G3V4']
ISOTOPE_LABELS = {
    'WT': 'UL', 'G3+': 'G1', 'G5+': 'G3', 'V1': 'V4', 'G3': 'G6', 
    'V4': 'V7', 'G5+G3': 'G3G6', 'V1V4': 'V4V7', 'V1G3V4': 'V4G6V7'
}

# Analysis Parameters
SHIFT_VALUES = [-2, 0, 2]
THETA_TARGET = 1.0
THETA_RANGE = np.logspace(-1, 4, 1000)

# Plotting Specifications
FIGURE_SIZE = (7, 3.5)  # Slightly wider to accommodate legend
DPI = 300
ISOTOPE_COLORS = [
    '#000000', '#E31A1C', '#1F78B4', '#33A02C', '#FF7F00',
    '#6A3D9A', '#FB9A99', '#A6CEE3', '#B2DF8A'
]
L_VALUE_RANGE = (0, 0.2)
KDE_POINTS = 200

# -----------------------------------------------------------------------------
# DATA LOADING
# -----------------------------------------------------------------------------

def load_error_data_for_isotope(isotope, force_field, theta_index, theta_value):
    """Load and process error data for specific isotope across all shifts."""
    all_l_values = []
    loaded_shifts = 0
    
    for shift in SHIFT_VALUES:
        filename = f'error_overlap_{isotope}_{force_field}_ch27_{shift}.npy'
        filepath = ERROR_DIR / filename
        
        if filepath.exists():
            error_data = np.load(filepath)
            # Calculate L values: error(theta) / theta
            l_values = error_data[:, theta_index] / theta_value
            all_l_values.extend(l_values)
            loaded_shifts += 1
        else:
            print(f'    Warning: Missing {filepath.name}')
    
    if loaded_shifts > 0:
        print(f'    {isotope}: loaded {loaded_shifts}/{len(SHIFT_VALUES)} shifts, '
              f'{len(all_l_values)} L-values')
    
    return all_l_values

# -----------------------------------------------------------------------------
# ANALYSIS
# -----------------------------------------------------------------------------

def find_target_theta_index(theta_range, target_theta):
    """Find index of theta value closest to target."""
    target_index = np.argmin(np.abs(theta_range - target_theta))
    actual_theta = theta_range[target_index]
    
    print(f'Target θ = {target_theta}, actual θ = {actual_theta:.6f} '
          f'(index {target_index})')
    
    return target_index, actual_theta

def calculate_kde_density(values, x_range, n_points):
    """Calculate kernel density estimation for given values."""
    if not values:
        return None, None
    
    # Create KDE
    kde = gaussian_kde(values)
    
    # Generate evaluation points within range
    x_min, x_max = x_range
    data_min, data_max = min(values), max(values)
    
    # Use intersection of specified range and data range
    eval_min = max(x_min, data_min)
    eval_max = min(x_max, data_max)
    
    if eval_min >= eval_max:
        return None, None
    
    x_eval = np.linspace(eval_min, eval_max, n_points)
    density = kde(x_eval)
    
    return x_eval, density

# -----------------------------------------------------------------------------
# PLOTTING
# -----------------------------------------------------------------------------

def set_plot_style():
    """Set consistent plot style for ACS publication figures."""
    plt.rcParams.update({
        # Font settings - Arial (PC) or Helvetica (Mac) as per ACS
        'font.family': 'sans-serif',
        'font.sans-serif': ['Arial', 'Helvetica'],
        'font.size': 10,  # Will be readable at 5pt when reduced to column width
        
        # Line widths - minimum 1 pt as per ACS
        'lines.linewidth': 1.0,
        'axes.linewidth': 1.0,
        'xtick.major.width': 1.0,
        'ytick.major.width': 1.0,
        'xtick.minor.width': 1.0,
        'ytick.minor.width': 1.0,
        'patch.linewidth': 1.0,
        'grid.linewidth': 1.0,
        
        # Font sizes - accounting for reduction
        'axes.labelsize': 11,
        'axes.titlesize': 12,
        'xtick.labelsize': 10,
        'ytick.labelsize': 10,
        'legend.fontsize': 10,
        
        # DPI settings
        'figure.dpi': 600
    })


def plot_kde_curve(ax, values, label, color):
    """Plot kernel density estimation curve for given values."""
    x_eval, density = calculate_kde_density(values, L_VALUE_RANGE, KDE_POINTS)
    
    if x_eval is not None and density is not None:
        ax.plot(x_eval, density, label=label, color=color, linewidth=1.0)
        return True
    else:
        print(f'    Warning: Could not plot KDE for {label}')
        return False

def setup_subplot_axes(ax, force_field, is_leftmost=False, is_rightmost=False):
    """Configure axes properties for consistent formatting."""
    ax.set_xlabel('L values', fontsize=10)
    
    if is_leftmost:
        ax.set_ylabel('Density', fontsize=10)
    elif is_rightmost:
        # For rightmost plot, still show y-axis ticks and labels
        ax.tick_params(axis='y', which='both', labelleft=True)
    
    ax.set_title(FORCE_FIELD_LABELS[force_field], fontsize=10)
    ax.set_xlim(L_VALUE_RANGE)
    ax.grid(True, alpha=0.3, linewidth=0.5)
    ax.tick_params(labelsize=8)

def create_kde_comparison_figure(theta_index, theta_value):
    """Create complete figure comparing L-value distributions."""
    # Create figure with better spacing
    fig, axes = plt.subplots(1, 2, figsize=FIGURE_SIZE, sharey=True, dpi=DPI)
    
    # Add some spacing between subplots
    plt.subplots_adjust(wspace=0.05)
    
    for ff_index, force_field in enumerate(FORCE_FIELDS):
        ax = axes[ff_index]
        
        print(f'\nProcessing {force_field}...')
        
        plotted_count = 0
        for isotope_index, isotope in enumerate(ISOTOPES):
            # Load L-values for this isotope
            l_values = load_error_data_for_isotope(
                isotope, force_field, theta_index, theta_value
            )
            
            # Plot KDE if data exists
            if l_values:
                isotope_label = ISOTOPE_LABELS[isotope]
                color = ISOTOPE_COLORS[isotope_index]
                
                success = plot_kde_curve(ax, l_values, isotope_label, color)
                if success:
                    plotted_count += 1
        
        print(f'  Successfully plotted {plotted_count}/{len(ISOTOPES)} isotopes')
        
        # Setup axes
        setup_subplot_axes(ax, force_field, 
                         is_leftmost=(ff_index == 0),
                         is_rightmost=(ff_index == 1))
    
    return fig, axes

def add_figure_legend(fig, axes):
    """Add legend to the right side of the figure."""
    # Get legend elements from first subplot
    handles, labels = axes[0].get_legend_handles_labels()
    
    # Add legend closer to the plots
    fig.legend(handles, labels, 
               bbox_to_anchor=(0.98, 0.5), loc='center left', 
               fontsize=9, frameon=True, framealpha=0.9,
               edgecolor='black', fancybox=False)

# -----------------------------------------------------------------------------
# MAIN
# -----------------------------------------------------------------------------

def main():
    """Main function to run L-values KDE analysis."""
    set_plot_style()
    
    print('L-values Kernel Density Estimation Analysis')
    print('=' * 50)
    
    try:
        # Find target theta index
        theta_index, actual_theta = find_target_theta_index(THETA_RANGE, THETA_TARGET)
        
        # Create KDE comparison figure
        fig, axes = create_kde_comparison_figure(theta_index, actual_theta)
        
        # Add legend
        add_figure_legend(fig, axes)
        
        # Adjust layout - use constrained layout for better automatic spacing
        plt.tight_layout()
        
        # Fine-tune spacing to accommodate legend without cutting off labels
        plt.subplots_adjust(right=0.88, left=0.08, bottom=0.15, top=0.92)
        
        plt.show()
        
        print('\nL-values KDE analysis completed successfully!')
        print('=' * 50)
        
    except Exception as e:
        print(f'Error during analysis: {e}')
        raise

if __name__ == '__main__':
    main()
```
