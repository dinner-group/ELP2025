```python
### DETERMINATION OF INFORMATIVE REGIONS FOR ALL 9 ISOTOPOLOGUES FOR BOTH FFs (FIG S15) ###
### 2 Aug 2025 ###

import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

# -----------------------------------------------------------------------------
# CONFIGURATION
# -----------------------------------------------------------------------------

# Paths (unified with Figure 8 structure)
BASE_DIR = Path('../data')
WAVENUMBERS_FILE = BASE_DIR / 'waxis.txt'
EXP_DIR = BASE_DIR / 'exp_data' / 'processed_FTIRs'
SIM_DIR = BASE_DIR / 'sim_data' / 'sim_spec'
PROP_DIR = BASE_DIR / 'prop_array'

# Data Specifications
FORCE_FIELD = 'CHARMM27_TIP3P'
MAP_TYPE = 'DC1Fnew'

# Isotope mapping (same as Figure 8)
ISOTOPE_MAP = {
    'G3':     {'label': 'G6',   'file': 'G3'},
    'G3+':    {'label': 'G1',   'file': 'G3dag'},
    'G5+':    {'label': 'G3',   'file': 'G5dag'},
    'G5+G3':  {'label': 'G3G6', 'file': 'G5dagG3'},
    'V1':     {'label': 'V4',   'file': 'V1'},
    'V1G3V4': {'label': 'V4G6V7', 'file': 'V1G3V4'},
    'V1V4':   {'label': 'V4V7', 'file': 'V1V4'},
    'V4':     {'label': 'V7',   'file': 'V4'},
    'WT':     {'label': 'UL',   'file': 'WT'},
}

# Analysis Parameters (same as Figure 8)
WAVENUMBER_RANGE = (1525, 1750)
MAX_WAVENUMBER = 1625
SIGMA = 0.0005
DIFF_THRESHOLD = 0.0015
WINDOW_WIDTH = 5
PURITY_THRESHOLD = 0.95

# Plotting Specifications
FIGURE_SIZE = (7, 9.5)
DPI = 600

# -----------------------------------------------------------------------------
# DATA LOADING (exact same functions as Figure 8)
# -----------------------------------------------------------------------------

def load_wavenumber_axis():
    """Load wavenumber axis within analysis range."""
    wavenumbers = np.loadtxt(WAVENUMBERS_FILE)
    return wavenumbers[(wavenumbers >= WAVENUMBER_RANGE[0]) & 
                       (wavenumbers <= WAVENUMBER_RANGE[1])]

def load_proportions():
    """Load proportions array for force field."""
    filepath = PROP_DIR / FORCE_FIELD / f'proportions_array_0100ps_{FORCE_FIELD}_PAPER.npy'
    if not filepath.exists():
        raise FileNotFoundError(f'Missing proportions file: {filepath}')
    return np.load(filepath)

def load_experimental(file_name, wavenumbers):
    """Load and normalize experimental spectrum."""
    filepath = EXP_DIR / f'GVGn1_{file_name}_50mgmL_phos_150mM_pH1_basecorr_processed.npy'
    if not filepath.exists():
        raise FileNotFoundError(f'Missing experimental file: {filepath}')
    spectrum = np.load(filepath)
    return normalize_by_area(spectrum, wavenumbers)

def load_simulated(name, wavenumbers):
    """Load and normalize simulated spectra."""
    filepath = SIM_DIR / FORCE_FIELD / f'{MAP_TYPE}_{FORCE_FIELD}_{name}_ftir.npy'
    if not filepath.exists():
        raise FileNotFoundError(f'Missing simulation file: {filepath}')
    raw_spectra = np.load(filepath)
    corrected = raw_spectra - np.min(raw_spectra, axis=1)[:, None]
    return np.array([normalize_by_area(spectrum, wavenumbers) for spectrum in corrected])

# -----------------------------------------------------------------------------
# ANALYSIS (exact same functions as Figure 8)
# -----------------------------------------------------------------------------

def normalize_by_area(spectrum, wavenumbers):
    """Normalize spectrum by integrated area."""
    area = np.trapz(spectrum, x=wavenumbers)
    return spectrum / area

def compute_state_averages(simulated, proportions, wavenumbers):
    """Compute average spectra for pure states."""
    averages = []
    for idx in range(4):  # C, E, NC, N
        mask = proportions[:, idx] > PURITY_THRESHOLD
        if np.any(mask):
            avg = normalize_by_area(np.mean(simulated[mask], axis=0), wavenumbers)
        else:
            avg = np.zeros_like(wavenumbers)
        averages.append(avg)
    return averages

def compute_diff_metrics(C, E, NC, N):
    """Compute difference metrics for C-turn and N-bend features."""
    diff_c = np.abs((C + NC) - (E + N))
    diff_n = np.abs((N + NC) - (E + C))
    return diff_c, diff_n

def compute_abs_diff(experimental, simulated_mean, wavenumbers):
    """Compute absolute difference between experimental and simulated."""
    return np.abs(experimental - normalize_by_area(simulated_mean, wavenumbers))

def compute_informativeness(diff, delta):
    """Compute informativeness metric."""
    return diff * np.exp(-delta**2 / (2 * SIGMA**2))

def find_informative_range(informativeness, wavenumbers):
    """Find wavenumber range where informativeness exceeds threshold."""
    mask = (wavenumbers < MAX_WAVENUMBER) & (informativeness > DIFF_THRESHOLD)
    indices = np.where(mask)[0]
    
    if len(indices) == 0:
        return None
    
    keep = []
    for idx in range(len(wavenumbers)):
        if wavenumbers[idx] < MAX_WAVENUMBER:
            if np.any(np.abs(wavenumbers[indices] - wavenumbers[idx]) <= WINDOW_WIDTH):
                keep.append(idx)
    
    if keep:
        return (wavenumbers[min(keep)], wavenumbers[max(keep)])
    return None

# -----------------------------------------------------------------------------
# PLOTTING
# -----------------------------------------------------------------------------

def set_plot_style():
    """Set consistent plot style for ACS publication figures."""
    plt.rcParams.update({
        # Font settings - Arial (PC) or Helvetica (Mac) as per ACS
        'font.family': 'sans-serif',
        'font.sans-serif': ['Arial', 'Helvetica'],
        'font.size': 10,  # Will be readable at 5pt when reduced to column width
        
        # Line widths - minimum 1 pt as per ACS
        'lines.linewidth': 1.0,
        'axes.linewidth': 1.0,
        'xtick.major.width': 1.0,
        'ytick.major.width': 1.0,
        'xtick.minor.width': 1.0,
        'ytick.minor.width': 1.0,
        'patch.linewidth': 1.0,
        'grid.linewidth': 1.0,
        
        # Font sizes - accounting for reduction
        'axes.labelsize': 11,
        'axes.titlesize': 12,
        'xtick.labelsize': 10,
        'ytick.labelsize': 10,
        'legend.fontsize': 10,
        
        # DPI settings
        'figure.dpi': 600
    })

def create_informativeness_plot(results, ranges_c, ranges_n, wavenumbers):
    """Create 9×2 grid plot showing informativeness analysis."""
    fig, axes = plt.subplots(9, 2, figsize=FIGURE_SIZE, sharex=True, sharey=True, dpi=DPI)
    
    # Make plots perfectly touching
    plt.subplots_adjust(wspace=0, hspace=0, left=0.08, right=0.98, top=0.96, bottom=0.06)
    
    # Add column titles
    axes[0, 0].set_title('N-bend Focus', fontsize=10, pad=5)
    axes[0, 1].set_title('C-turn Focus', fontsize=10, pad=5)
    
    # Define order to match Figure 8's approach
    isotope_names = ['WT', 'G3+', 'G5+', 'V1', 'G3', 'V4', 'G5+G3', 'V1V4', 'V1G3V4']
    
    for i, name in enumerate(isotope_names):
        if name not in results:
            continue
            
        d = results[name]
        label = ISOTOPE_MAP[name]['label']
        
        # N-bend plot (left column)
        ax_n = axes[i, 0]
        info_n = compute_informativeness(d['diff_n'], d['delta'])
        
        ax_n.plot(wavenumbers, d['diff_n'], 'g:', lw=1.0, label=r'$I_{\mathrm{diff}}(\omega)$')
        ax_n.plot(wavenumbers, d['delta'], 'r--', lw=1.0, label=r'$\delta(\omega)$')
        ax_n.plot(wavenumbers, info_n, 'purple', lw=1.2, label=r'$D(\omega)$')
        ax_n.axhline(DIFF_THRESHOLD, color='gray', ls='--', lw=1.0)
        
        # Highlight informative region
        if name in ranges_n:
            ax_n.axvspan(*ranges_n[name], color='lightgreen', alpha=0.3)
        
        ax_n.text(0.02, 0.88, label, transform=ax_n.transAxes, fontsize=8, va='top')
        
        # Only show x-axis ticks on bottom row
        if i < 8:
            ax_n.tick_params(axis='x', which='both', bottom=False, labelbottom=False)
            ax_n.tick_params(axis='y', labelsize=8)
        else:
            ax_n.tick_params(axis='both', labelsize=8)
            ax_n.set_xlabel(r'$\omega$ (cm$^{-1}$)', fontsize=10)
        
        if i == 0:
            ax_n.legend(fontsize=8, loc='upper right')
        
        # C-turn plot (right column)
        ax_c = axes[i, 1]
        info_c = compute_informativeness(d['diff_c'], d['delta'])
        
        ax_c.plot(wavenumbers, d['diff_c'], 'b:', lw=1.0)
        ax_c.plot(wavenumbers, d['delta'], 'r--', lw=1.0)
        ax_c.plot(wavenumbers, info_c, 'purple', lw=1.2)
        ax_c.axhline(DIFF_THRESHOLD, color='gray', ls='--', lw=1.0)
        
        # Highlight informative region
        if name in ranges_c:
            ax_c.axvspan(*ranges_c[name], color='lightblue', alpha=0.3)
        
        ax_c.text(0.02, 0.88, label, transform=ax_c.transAxes, fontsize=8, va='top')
        ax_c.set_yticks([])
        
        # Only show x-axis ticks on bottom row
        if i < 8:
            ax_c.tick_params(axis='x', which='both', bottom=False, labelbottom=False)
        else:
            ax_c.tick_params(axis='x', labelsize=8)
            ax_c.set_xlabel(r'$\omega$ (cm$^{-1}$)', fontsize=10)
    
    # Add common y-label
    #fig.text(0.04, 0.5, 'Normalized Absorbance', rotation=90, 
    #        fontsize=10, ha='center', va='center')
    
    return fig

# -----------------------------------------------------------------------------
# MAIN
# -----------------------------------------------------------------------------

def main():
    """Main function to run informativeness analysis."""
    set_plot_style()
    
    print('Informativeness Analysis for Supplementary Information')
    print('=' * 60)
    
    try:
        # Load data
        wavenumbers = load_wavenumber_axis()
        proportions = load_proportions()
        
        print(f'Loaded wavenumber axis: {len(wavenumbers)} points')
        print(f'Range: {wavenumbers.min():.1f} - {wavenumbers.max():.1f} cm⁻¹')
        print(f'Loaded proportions array: {proportions.shape}')
        print()
        
        # Process all isotopes (using same logic as Figure 8)
        results = {}
        ranges_c, ranges_n = {}, {}
        
        for name, info in ISOTOPE_MAP.items():
            try:
                print(f'Processing {name}...')
                
                # Load data
                experimental = load_experimental(info['file'], wavenumbers)
                simulated = load_simulated(name, wavenumbers)
                C, E, NC, N = compute_state_averages(simulated, proportions, wavenumbers)
                
                # Compute metrics
                delta = compute_abs_diff(experimental, np.mean(simulated, axis=0), wavenumbers)
                diff_c, diff_n = compute_diff_metrics(C, E, NC, N)
                
                # Find informative ranges
                D_c = compute_informativeness(diff_c, delta)
                D_n = compute_informativeness(diff_n, delta)
                range_c = find_informative_range(D_c, wavenumbers)
                range_n = find_informative_range(D_n, wavenumbers)
                
                # Store results
                results[name] = {
                    'experimental': experimental, 'C': C, 'E': E, 'NC': NC, 'N': N,
                    'delta': delta, 'diff_c': diff_c, 'diff_n': diff_n
                }
                
                if range_c:
                    ranges_c[name] = range_c
                if range_n:
                    ranges_n[name] = range_n
                    
            except FileNotFoundError as e:
                print(f"  ERROR: {e}")
        
        print()
        print('Creating informativeness plot...')
        fig = create_informativeness_plot(results, ranges_c, ranges_n, wavenumbers)
        
        # Print comprehensive summary of informative regions
        print("\n" + "="*70)
        print("INFORMATIVE REGIONS SUMMARY")
        print("="*70)
        
        # C-turn regions
        print("\nC-TURN INFORMATIVE REGIONS:")
        if ranges_c:
            for name in sorted(ranges_c.keys()):
                start, end = ranges_c[name]
                label = ISOTOPE_MAP[name]['label']
                width = end - start
                print(f"  {name:<8} ({label:<8}): {start:6.1f} - {end:6.1f} cm⁻¹  (width: {width:4.1f} cm⁻¹)")
        else:
            print("  None found")
        
        # N-bend regions
        print("\nN-BEND INFORMATIVE REGIONS:")
        if ranges_n:
            for name in sorted(ranges_n.keys()):
                start, end = ranges_n[name]
                label = ISOTOPE_MAP[name]['label']
                width = end - start
                print(f"  {name:<8} ({label:<8}): {start:6.1f} - {end:6.1f} cm⁻¹  (width: {width:4.1f} cm⁻¹)")
        else:
            print("  None found")
        
        print("="*70 + "\n")
        
        plt.show()
        
        print('Informativeness analysis completed successfully!')
        
    except Exception as e:
        print(f'Error during analysis: {e}')
        raise

if __name__ == '__main__':
    main()
```
